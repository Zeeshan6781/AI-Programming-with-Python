```python
# Making necessary imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Loading data into a dataset
fuel_econ = pd.read_csv("fuel-econ.csv")
fuel_econ.head(5)
```

```python
base_color = sns.color_palette()[0]
sns.barplot(x = "VClass", y = "comb", data = fuel_econ, color = base_color)
plt.xticks(rotation = 15)
plt.ylabel('Avg. Combined Fuel Efficiency (mpg)');
```

```python
sns.barplot(x = "VClass", y = "comb", data = fuel_econ, color = base_color, errwidth = 0);
plt.xticks(rotation = 15);
```

```python
sns.barplot(x = "VClass", y = "comb", data = fuel_econ, color = base_color, ci = "sd");
plt.xticks(rotation = 15);
```

```python
sns.pointplot(x = "VClass", y = "comb", data = fuel_econ, color = base_color, linestyles = "");
plt.xticks(rotation = 15)
plt.ylabel('Avg. Combined Fuel Efficiency (mpg)');
```

## Example 3. Bringing a few charts together

```python
plt.figure(figsize = [20, 5])
base_color = sns.color_palette()[0]

# left plot: violin plot
plt.subplot(1, 3, 1)
sns.violinplot(x = "VClass", y = "comb", data = fuel_econ, inner = None, color = base_color)
plt.xticks(rotation = 45)

# center plot: box plot
plt.subplot(1, 3, 2)
sns.boxplot(x = "VClass", y = "comb", data = fuel_econ, color = base_color)
plt.xticks(rotation = 45)

# right plot: adapted bar chart
plt.subplot(1, 3, 3)
sns.barplot(x = "VClass", y = "comb", data = fuel_econ, color = base_color)
plt.xticks(rotation = 45);
```
