## Default heat plot using Matplotlib.pyplot.hist2d() function

```python
# Making necessary imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Making necessary imports
fuel_econ = pd.read_csv("fuel-econ.csv")
fuel_econ.head(5)
```

```python
plt.figure(figsize = [20, 5])

# PLOT ON LEFT
plt.subplot(1, 2, 1)
sns.regplot(x = "displ", y = "comb", data = fuel_econ, x_jitter = 0.04, scatter_kws = {"alpha" : 1/10}, fit_reg = False)
plt.xlabel("Displacement (1)")
plt.ylabel("Combined Fuel Eff. (mpg)")

# PLOT ON RIGHT
plt.subplot(1, 2, 2)
plt.hist2d(x = "displ", y = "comb", data = fuel_econ)
plt.colorbar()
plt.xlabel("Displacement (1)")
plt.ylabel("Combined Fuel Eff. (mpg)");
```

## Heat plot - Set a minimum bound on counts and a reverse color map

```python
# Use cmin to set a minimum bound of counts 
# Use cmap to reverse the color map. 
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin=0.5, cmap='viridis_r')
plt.colorbar()
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Eff. (mpg)');
```

## Heat plot - Specify bin edges

```python
# Specify bin edges
bins_x = np.arange(0.6, 7+0.3, 0.3)
bins_y = np.arange(12, 58+3, 3)

plt.hist2d(x = "displ", y = "comb", data = fuel_econ, cmin = 0.5, cmap = "viridis_r", bins = [bins_x, bins_y])
plt.colorbar()
plt.xlabel("Displacement (1)")
plt.ylabel("Combined Fuel Eff. (mpg)");
```
